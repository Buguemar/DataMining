## 4. Aplicación de una red neuronal en Localización
Desde la edad antigua, múltiples formas de localización han sido desarrolladas. Dentro de los avances más importantes en este ámbito, es el desarrollo de la teorı́a cientı́fica y técnica denominada georreferenciación. Gracias a GPS, el crecimiento y acceso de la georreferenciación y navegación está en progresivo aumento, el problema surge cuando se intentan estimar en recintos interiores (como edificios o bajo tierra) donde el GPS no funciona de la manera como uno esperaría, debido a que existen muchos obstáculos e interferencia que imposibilitan su uso.  
Dentro de interiores se cuenta con señales RSSI (*fingerprint*) que pueden atacar este problema, sin embargo los métodos actuales no son robustos a ruido, por lo que su tarea será la de abordar este problema para mejorar exactitud de sistemas de posicionamiento en interiores mediante redes neuronales.  

La metodología con la que se trabajará será que, para dentro de interiores, dispositivos *Bluetooth* emiten señales RSSI las cuales son captadas por el dispositivo "objetivo" al cual se le desea determinar su localización, recibiendo distintas intensidades de señal de cada dispositivo emisor debido a su posición en el interior. Los datos con los que se va a trabajar (*IndoorFingerprint.csv*) fueron provistos por el nuevo Ing. Civil Informático Felipe Berrios, éstos constan de 8 características (*C1hA,0kxZ,tvMX,OlYb,7rk5,F39L,VNSF,tkxI*) correspondientes a las mediciones hechas/recibidas por el dispositivo "objetivo" de las distintas señales RSSI emitidas por los dispositivos *Bluetooth* en los bordes del interior, además de tener la posición del dispositivo "objetivo" en un plano XY (valor a estimar).

<img src="https://i.imgur.com/Xheipaa.png" width="60%" height="20%" />

Grilla ejemplo de cómo funciona el sistema (elaboración por Felipe Berrios). Los 4 dispositivos en la esquina son los que emiten las señales RSSI, el punto naranja es el dispositivo que las recibe y es el "objetivo" a determinar la posición.

Una consideración importante es el cómo tratar la ausencia de la señal proveniente de un dispositivo *Bluetooth*, para estos datos se utiliza un valor de +100, ya que es imposible obtener este valor debido a las características de la escala RSSI (siempre negativa o igual a cero), pero puede ser sustituido por otro.  
Para hacer el trabajo mas simple se discretizará la posición en el plano definiendo zonas en dónde está el objeto a localizar. Las zonas deben ser las que indica la malla a continuación:

````python
import pandas as pd
import numpy as np

df = pd.read_csv("./IndoorFingerprint.csv")

import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df["X"],df["Y"])

x_ticks = np.arange(0, 49, 8)
y_ticks = np.arange(5, 22, 4)
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.grid(color='r', linestyle='-', linewidth=2)
plt.ylabel("y position")
plt.xlabel("x position")
plt.show()
````

Donde los puntos azules son los distintos datos superpuestos de las posiciones del objeto a localizar. Por ejemplo el punto (2,8) está en la primera zona (o en la primera zona del eje *x* y del eje *y*), el punto (2,20) está en la zona 19 (o en la primera zona del eje *x* y la cuarta zona del eje *y*). Esta discretización transforma el problema que en un principio podría ser de regresión para determinar la posición exacta, en un problema de clasificación dividiendo (dentro de los posibles valores) 6 zonas para el eje "*x*" y 4 zonas para el eje "*y*", contando con un total de 24 clases (24 zonas en la malla).

> Deberá entrenar una red neuronal *feed forward* para la clasificación de las 24 posibles clases, con el objetivo de tener un desempeño (*accuracy*) mayor al 75%.

En las celdas anteriores se aprecian los primeros 5 registros del dataframe a utilizar en la tarea de predicción de zonas dentro de una grilla tal como la graficada anteriormente. Igualmente, es posible notar que el dataframe se compone de 6600 datos, cada uno descrito por 10 atributos (indicados en la celda inmediatamente anterior).  
Dado que, como ya se enunció, la tarea de predicción se ha discretizado, es necesario definir una nueva columna ('target') la cual indique la celda que se corresponda con los valores 'X' e 'Y' del registro en cuestión. 

````python
def to_celda(X,Y):
    horizontal= np.ceil(X/8)
    vertical=np.ceil(Y/4)-1
    hor,vert= [np.int(horizontal),np.int(vertical)]
    return hor+6*(vert-1)    

celda=[]
for i,j in zip(df['X'],df['Y']):
    celda.append(to_celda(i,j))

df['Celda']= pd.DataFrame(celda)
df.shape
````

Hecho esto, otro punto que cabe mencionar antes de proceder con la labor propuesta, corresponde al análisis estadístico de los datos; Esto es, rango de valores que toman los datos, sus valores máximos/mínimos, promedio, desviación estándar, etc. Todo esto, podría tener efectos no deseados en el entrenamiento del learner a establecer pudiendo priorizar determinados atributos, o bien, empeorar el proceso de aprendizaje de la máquina.  
Así entonces, es notable la gran varianza de algunos atributos, lo que se hace que los datos se ubiquen en rangos numéricos bastante dispersos. Por ejemplo, 'tkxI' vs 'Y'

### Inicialización uniforme, activación ReLU, optimizador SGD

````python
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled2.shape[1], kernel_initializer='uniform',activation="relu"))
model.add(Dense(16, kernel_initializer='uniform',activation="relu"))
model.add(Dense(24, kernel_initializer='uniform',activation="softmax")) 
model.compile(optimizer=SGD(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled2, y_onehot, batch_size=20, epochs=200,verbose=1,validation_split=0.2)#, validation_data=(X_test_scaled, y_test))
````

````python
print(history.history.keys())


import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14,8))
plt.plot(history.history['acc'], label='Train')
plt.plot(history.history['val_acc'], label='Val')
plt.title('Uniform + ReLU, accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
````

Para la configuración más básica, se aprecia la lenta convergencia hacia los valores óptimos. Así, desde el epoch 75 se alcanzan valores significativos de accuracy (sobre el 80%).  
Cabe destacar que en este modelo no se hizo uso de ningún regularizador ni tampoco de alguna técnica de optimización moderna como las estudiadas en la pregunta 1. 

Por ello, se procede a aplicar regularizador L1 (en la primera capa oculta y la última), además del optimizador Adam, el cual demostró ser bastante competente dadas las experimentaciones anteriores en esta tarea.

### Inicialización uniforme, activación ReLU, regularización L1, optimizador Adam

````python
import keras
from keras.regularizers import *
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD, Adam
y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled2.shape[1], kernel_initializer='uniform', W_regularizer=l1(0.0001),activation="relu"))
model.add(Dense(16, kernel_initializer='uniform',activation="relu"))
model.add(Dense(24, kernel_initializer='uniform', W_regularizer=l1(0.001),activation="softmax")) 
model.compile(optimizer=Adam(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled2, y_onehot, batch_size=20, epochs=200,verbose=1,validation_split=0.2)#, validation_data=(X_test_scaled, y_test))
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14,8))
plt.plot(history.history['acc'], label='Train')
plt.plot(history.history['val_acc'], label='Val')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
````

Esta vez se aprecia una convergencia mucho más rápida hacia los valores óptimos. Sin embargo, la variabilidad  en términos del accuracy es notable, por lo que, a pesar de cumplir con lo solicitado (accuracy sobre 75%), no se aceptará como el mejor modelo entrenado y se buscará una mejor configuración. 


### Inicialización uniforme de Glorot, activación Sigmoid, regularización L1 y Dropout, optimizador Adam

````python
import keras
from keras.regularizers import *
from keras.models import Sequential
from keras.layers.core import Dense, Activation,Dropout
from keras.optimizers import SGD, Adam

y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled2.shape[1], kernel_initializer='glorot_uniform', W_regularizer=l1(0.0001),activation="sigmoid"))
model.add(Dense(16, kernel_initializer='glorot_uniform',activation="sigmoid"))
model.add(Dropout(0.001))
model.add(Dense(24, kernel_initializer='glorot_uniform', W_regularizer=l1(0.001),activation="softmax")) 
model.compile(optimizer=Adam(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled2, y_onehot, batch_size=20, epochs=200,verbose=1,validation_split=0.2)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14,8))
plt.plot(history.history['acc'], label='Train')
plt.plot(history.history['val_acc'], label='Val')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
````

En esta nueva configuración es clara la gran mejora que se ha obtenido, bastan unos pocos epochs para converger hacia los valores óptimos y desaparece casi por completo el efecto de varianza en las predicciones. Es decir, no hay fluctuaciones significativas en términos de accuracy tanto en el conjunto de entrenamiento como en el validación.


### Inicialización uniforme de Glorot, activación Sigmoid, regularización Dropout, optimizador Adam

Ahora, se presenta unaa nueva configuración que elimina la aplicación de regularzadores L1 a modo de simlificar el modelo. Con esto se obtiene: 

````python
import keras
from keras.regularizers import *
from keras.models import Sequential
from keras.layers.core import Dense, Activation,Dropout
from keras.optimizers import SGD, Adam

y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled2.shape[1], kernel_initializer='glorot_uniform', activation="sigmoid"))
model.add(Dense(16, kernel_initializer='glorot_uniform',activation="sigmoid"))
model.add(Dropout(0.001))
model.add(Dense(24, kernel_initializer='glorot_uniform',activation="softmax")) 
model.compile(optimizer=Adam(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled2, y_onehot, batch_size=20, epochs=200,verbose=1,validation_split=0.2)
````

Se aprecia entonces que no existen diferencias notables en el rendimiento del modelo entrenado respecto al anterior. Sin embargo, existe una leve varianza en los valores de accuracy pero aún así no logra ser significativa. Esto se puede deber a la eliminación de los regularizadores L1. 


### Inicialización uniforme de He, activación ReLU, regularizador Dropout, optimizador Adam

A continuación se emplea la misma configuración anterior pero utilizando función de activación ReLU e inicialización de pesos uniforme de He, la cual aplica el mismo enfoque que utilizó Glorot para la activación  Sigmoidal.
````python
y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled2.shape[1], kernel_initializer='he_uniform', activation="relu"))
model.add(Dense(16, kernel_initializer='he_uniform',activation="relu"))
model.add(Dropout(0.001))
model.add(Dense(24, kernel_initializer='he_uniform', activation="softmax")) 
model.compile(optimizer=Adam(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled2, y_onehot, batch_size=30, epochs=200,verbose=1,validation_split=0.2)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(14,8))
plt.plot(history.history['acc'], label='Train')
plt.plot(history.history['val_acc'], label='Val')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
````

Por ello, es intuitivo que la varianza en el modelo será más pronunciada que utilizando, o un dropout menor, o bien, un optimizador más lento.

### Inicialización uniforme de He, activación ReLU, regularización Dropout, optimizador SGD

En esta nueva configuración, se propone mantener la estructura y optimizador pero se disminuye la tasa de dropout, a fin de que el modelo cuente con mayor información durante el entrenamiento.

````python
y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
model = Sequential()
model.add(Dense(32, input_dim=X_train_scaled.shape[1], kernel_initializer='he_uniform', activation="relu"))
model.add(Dense(16, kernel_initializer='he_uniform',activation="relu"))
model.add(Dropout(0.01))
model.add(Dense(24, kernel_initializer='he_uniform', activation="softmax")) 
model.compile(optimizer=SGD(lr=0.01),loss='categorical_crossentropy', metrics=["accuracy"])
history = model.fit(X_train_scaled, y_onehot, batch_size=30, epochs=200,verbose=1,validation_split=0.2)
````

````python
plt.figure(figsize=(14,8))
plt.plot(history.history['acc'], label='Train')
plt.plot(history.history['val_acc'], label='Val')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
````

Si bien la tendencia del accuracy parece no variar, no acelerandose ni ralentizadose el aprendizaje, parece ser que esta configuración logra mejor generalización que el modelo previo. Esto es, los valores de accuracy para el conjunto de validación (en naranja) superan levemente al accuracy del conjunto de entrenamiento.

````python
from keras.optimizers import SGD, Adam, RMSprop
y_onehot = keras.utils.to_categorical(y_train)
y_onehot= y_onehot[:,1:]
modelHe = Sequential()
modelHe.add(Dense(128, kernel_initializer='he_uniform', input_dim=X_train_scaled2.shape[1],activation="relu"))
modelHe.add(Dense(64, kernel_initializer='he_uniform', activation="relu"))
modelHe.add(Dropout(0.03))
modelHe.add(Dense(24, activation="softmax")) 
modelHe.compile(optimizer='adam',loss='categorical_crossentropy', metrics=["accuracy"])
history = modelHe.fit(X_train_scaled2, y_onehot, epochs=200,verbose=1,validation_split=0.2,batch_size=30)
````
